```python
# para saber si existe "media" en los keys de l['entities_str'] - Sol 1
lista4 = [{'id_str': l['id_str'], 'entities_str': l['entities_str']} for l in lista3 \
          if 'media' in l['entities_str']]

# para saber si existe "media" en los keys de l['entities_str'] - Sol 2
lista4 = []
for l in lista3:
    if 'media' in l['entities_str'].keys():
        lista4.append({'id_str': l['id_str'], 'entities_str': l['entities_str']})
```

```python
# uni√≥n de la id_str con cada posici√≥n de listado de entities
def manejo_lista(id, lista):
    return [(id, l['text']) for l in lista]
```

```python
# usando map
lista_hastags = list(map(lambda x: (manejo_lista(x['id_str'], x['entities_str']['hashtags'])), lista3))
print(lista_hastags)
```

```python
# funci√≥n para obtener una lista y guardar en el archivo
def archivo_hashtags(lista):
    archivo = open("data/hashtags_id.csv", "a")
    for l in lista:
        archivo.write("%s|%s\n" % (l[0], l[1]))
    archivo.close()
    return True
```

```python
# guardar en el archivo
[archivo_hashtags(l) for l in lista_hastags]
```
